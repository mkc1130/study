```python
import pandas as pd

pd.set_option('mode.chained_assignment',  None)

df = pd.read_csv('~/dt3.csv')

```

```python
dis_pm2 = []
for i in range(0,len(df)) :
    if df['dis_pm2.5'][i] == df['pm2.5'][i] : dis_pm2.append(np.nan)
    else : dis_pm2.append(df['dis_pm2.5'][i])
        
dis_pm10 = []
for i in range(0,len(df)) :
    if df['dis_pm10'][i] == df['pm10'][i] : dis_pm2.append(np.nan)
    else : dis_pm10.append(df['dis_pm10'][i])
```

```python

pm2 = pd.to_numeric(df['pm2.5'])
dis_pm2 = pd.to_numeric(df['dis_pm2.5'])

pm10 = pd.to_numeric(df['pm10'])
dis_pm10 = pd.to_numeric(df['dis_pm10'])
```
